```python
# libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#df = sns.load_dataset('tested')
#df.head()
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
# scatter plot
sns.scatterplot(data=df, x='age',y='fare',hue='class')
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='age',y='fare', hue='survived')
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='age',y='fare',hue='survived',palette='viridis',sizes=(20,200),alpha=0.5)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x='age',
    y='fare',
    hue='survived',
    palette='dark',
    sizes=(20,200),
    alpha=0.5
)
plt.title('Age vs fare by survival')
plt.show()
```

```python
# using dark background for plotting the scatter plot
plt.style.use('dark_background')
sns.set_palette('dark')
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x='age',
    y='fare',
    hue='survived',
    palette='dark',
    sizes=(20,200),
    alpha=1.0
)
plt.title('Age vs Fare survival on titanic')
plt.show()
```

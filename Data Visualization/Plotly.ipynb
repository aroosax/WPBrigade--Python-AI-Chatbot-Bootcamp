```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('titanic') # load data
```

```python
# clean data: drop all rows with NaNs
df.dropna(how='all', inplace=True)
```

```python
# 1. survival count
fig1 = px.histogram(df, x='survived', title='Survival count(0 = No , 1 = Yes)', color='survived')
fig1.show()
```

```python
# 2. Gender distributiom
fig2 = px.pie(df, names='sex', title='Gender distribbution')
fig2.show()
```

```python
# 3. survived by gender
fig3 = px.histogram(df, x='sex', color='survived', barmode='group',title='Survived by Gender')
fig3.show()
```

```python
# 4. passenger class distribution
fig4 = px.pie(df, names='class',title='Passenger class distribution')
fig4.show()
```

```python
# 5. survival by passneger class
fig5 = px.histogram(df, x='class',color='survived',barmode='group',title='Survival by the class of the passenger')
fig5.show()
```

```python
# 6. Age distribution
fig6 = px.histogram(df, x='age',nbins=30, title='Age distribution')
fig6.show()
```

```python
# 7. Age vs Fare (colored by survival)
fig7 = px.scatter(df, x='age',y='fare',color='survived',title='Age vs Fare (colored by survival)')
fig7.show()
```

```python
# 8. Age vs Fare (colored by class)
fig8 = px.scatter(df, x='age',y='survived',color='class',symbol='sex', title='Age vs Fare (colored by class)')
fig8.show()
```

```python
# 9. Box plot of fare by class
fig9 = px.box(df, x='class',y='fare', color='class',title='Box plot of fare by class')
fig9.show()
```

```python
# 10. Box plot of age by gender
fig10 = px.box(df, x='sex',y='age',color='sex',title='Box plot of age by gender')
fig10.show()
```

```python
# 11. survival by embrak town
fig11 = px.histogram(df, x='embark_town',color='survived',barmode='group',title='survival by embark town')
fig11.show()
```

```python
# 12.Heatmap correlation
numeric_df = df[['age', 'fare', 'pclass', 'parch', 'sibsp', 'survived']].dropna()
fig12 = px.imshow(numeric_df.corr(),text_auto=True, title='Heatmap correlation')
fig12.show()
```

```python
# 13. average fare by class
avg_fare = df.groupby('class')['fare'].mean().reset_index()
fig13 = px.bar(avg_fare, x='class',y='fare',color='class',title='Average fare by class')
fig13.show()
```

```python
# 14.Survival by Age Bins
df['age_bin'] = pd.cut(df['age'], bins=[0, 12, 20, 40, 60, 80])
# Convert Interval to string for plotting
df['age_bin_str'] = df['age_bin'].astype(str)
fig14 = px.histogram(df, x='age_bin_str',color='survived',barmode='group',title='Survival by age bins')
fig14.show()
```

```python
# 15. Strip plot: Age vs Survival by Gender
fig15 = px.strip(df, x='survived',y='age',color='sex',title='Strip plot: Age vs Survival by Gender')
fig15.show()
```

```python
# 16.sunburst chart
fig16 = px.sunburst(df,
                    path=['sex','class','survived'],
                    color='survived',
                    title='Survival Breakdown by Gender and Class'
                    )
fig16.show()
```

```python
# 17. Animated scatter plot
fig17 = px.scatter(df,
                   x='age',
                   y='fare',
                   color='survived',
                   animation_frame='class',
                   title='Age vs Fare Animated by Passenger Class',
                   size='fare',
                   hover_name='sex',
                   range_x=[0,80],
                   range_y=[0,500]
                   )
fig17.show()
```

```python
# 18. Animated Bar Chart
fig18 = px.histogram(df,
                     x='embark_town',
                     color='survived',
                     animation_frame='class',
                     barmode='group',
                     title='Survival by Embark Town Animated by Class'
                     )
fig18.show()
```

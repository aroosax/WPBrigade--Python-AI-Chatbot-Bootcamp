```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# ðŸ“Š Matplotlib

Matplotlib is a powerful Python library for data visualization.

It allows us to create plots like line charts, bar charts, scatter plots, and more.

It is highly customizable, making it useful for both simple and complex visualizations.

# ðŸŽ¨ Seaborn

Seaborn is built on top of Matplotlib and makes visualizations more attractive and easier to create.

It comes with built-in themes and color palettes for better styling.

It is especially useful for statistical plots like heatmaps, pair plots, and distributions.

```python
# load iris dataset
df = sns.load_dataset('iris')
df.head()
```

```python
# histogram of sepal_length from iris dataset
plt.hist(df['sepal_length'], bins = 10, color = 'red' ,alpha = 0.6) # alpha shows the darkness of the color red
plt.title("Histogram of sepal_length")
plt.xlabel("sepal_length")
plt.ylabel('Frequency')
plt.show()
```

```python
# box plot for all the numeric columns in the iris dataset
sns.boxplot(data=df)
plt.title("Box plot of iris featues")
plt.show()
```

```python
print(df.describe())
```

```python
# scatter plot with different species with different colours
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species', palette='Set2', s=100)
plt.title("Scatter plot: sepal_length vs petal_length")
plt.xlabel("Petal_Length")
plt.ylabel("Sepal_Length")
plt.legend(title='Species')


plt.show()
```

```python
# line chart
plt.plot(df['sepal_length'], label='Sepal_Length')
plt.plot(df['petal_length'], label='Petal_length')
plt.title("Line chart of Sepal_Length and Petal_length")
plt.xlabel("Index")
plt.ylabel("Length (cm)")
plt.legend()
plt.show()
```

```python
# pie chart for species distribution
species_count = df['species'].value_counts()
plt.pie(species_count.values, labels=species_count.index, autopct='%1.1f%%', startangle=140)
#plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%', startangle=140)
plt.title("Species Distribution")
plt.show()
```

```python
# donut chart for species distribution
plt.pie(species_count.values, labels=species_count.index, autopct='%1.1f%%', startangle=140, wedgeprops={'width':0.4})
plt.title("Donut chart showing species distribution")
plt.show()
```

```python
# pair plot for iris dataset
sns.pairplot(df, hue='species')
plt.show()
```

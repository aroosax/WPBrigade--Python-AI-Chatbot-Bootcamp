# Visualizing data in Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Ice Cream Ratings.csv')
df.set_index('Date', inplace=True)
df
```

```python
# styling in plt
print(plt.style.available)
plt.style.use('ggplot')
```

```python
# by default it is a line plot
#df.plot()
df.plot(kind='line', title='Ice Cream Rating', xlabel='Daily Rating',ylabel='Score')
```

```python
# if i want 3 different line graphs
df.plot(kind='line',subplots=True)
```

```python
# bar plot
df.plot(kind='bar',title='Ice Cream Rating',xlabel='Daily Rating',ylabel='Score')
```

```python
# stacked bar chart
df.plot(kind='bar',stacked=True, title='Ice Cream Rating',xlabel='Dailt Rating',ylabel='Scores')
```

```python
# bar chart showing only one element
df['Flavor Rating'].plot(kind='bar',title='Ice Cream Rating',xlabel='Daily Rating',ylabel='Scores')
```

```python
# horizontal barchart
df.plot(kind='barh', title='Ice Cream Rating',xlabel='Daily Rating',ylabel='Scores')
```

```python
# scatter plot
df.plot.scatter(x='Texture Rating',
                y='Overall Rating',
                s= 100)
```

```python
# histogram
df.plot.hist(bins=10)
```

```python
# boxplot
df.plot.box()
```

```python
# area plot
df.plot.area(figsize=(10,5))
```

```python
# pie chart
df.plot.pie(y='Flavor Rating',figsize=(10,5))
```

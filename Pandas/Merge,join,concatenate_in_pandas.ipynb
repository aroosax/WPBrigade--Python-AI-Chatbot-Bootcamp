# Merge, Join, anc Concatenate

```python
import pandas as pd
```

```python
# dataframe 1
df1 = pd.read_csv('LOTR.csv')
df1
```

```python
# dataframe 2
df2 = pd.read_csv('LOTR 2.csv')
df2
```

# Merge

```python
# inner join (df1.merge(df2)) by only writing this default it becomes inner join
df1.merge(df2, how='inner',on=['FellowshipID','FirstName'])
```

```python
# outer join (it gives all the data from df1 and df2)
df1.merge(df2, how='outer')
```

```python
# left join (connects all the values from the left table to the matching values of the right table)
df1.merge(df2, how='left')
```

```python
# right join (connects all the values of the right table to the matching values of the left table)
df1.merge(df2, how='right')
```

```python
# cross join (take one value from df1 and compare it to the all values of df2 it's also called as the cartesian product)
df1.merge(df2, how='cross')
```

# Join

```python
# outer join
df1.join(df2, on='FellowshipID', how='outer',lsuffix='_Left',rsuffix='_Right')
```

```python
# joins use indexes to work better
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix='_Left',rsuffix='_Right',how='outer')
df4
```

# Concatenate
It just puts one on top of the other

```python
# concatenation of 2 dataframes
pd.concat([df1,df2])
```

```python
# concatenate using joins
pd.concat([df1,df2],join='inner', axis=1)
```

# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel('Customer Call List.xlsx')
df
```

```python
# removing duplicates
df = df.drop_duplicates()
df
```

```python
# dropping columns
df = df.drop(columns='Not_Useful_Column')
df
```

```python
# cleaning data using strip (this is the first way)
#df['Last_Name'] = df['Last_Name'].str.lstrip('/')
#df['Last_Name'] = df['Last_Name'].str.lstrip('...')
#df['Last_Name'] = df['Last_Name'].str.rstrip('_')
# this is the second way
df['Last_Name'] = df['Last_Name'].str.strip('123./_')
df
```

```python
# making the phone number column standardized
#df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True) # this removes all the -/
#df['Phone_Number'].apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:10]) # this does not work due to data type float
#df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x)) ## converts datatype to string
#df['Phone_Number'] = df['Phone_Number'].apply(lambda x:x[0:3]+'-'+x[3:6]+'-'+x[6:10]) # puts - according to indexes
#df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
df
```

```python
# splitting columnns (adress column)
df[['Street_Adress','State','Zip_Code']] = df['Address'].str.split(',',n =2, expand=True)
df
```

```python
# standardizing columns (Paying customers)
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df
```

```python
# standardizing columns (Do_Not_Contact)
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
# showing all null(NaN) values empty
df = df.fillna('')
df
```

```python
# replacing all N/a with nothing
df = df.replace('N/a','')
df
```

```python
# filrering rows of data like if someone says yes do not contact then we remove that entire row
for x in df.index:
  if df.loc[x, 'Do_Not_Contact'] == 'Y':
    df.drop(x, inplace=True)

df
```

```python
# now drop the entire rowa where there is no phone number
for x in df.index:
  if df.loc[x, 'Phone_Number'] == '':
    df.drop(x, inplace=True)

df

# another way to drop null values
#df = df.dropna(subset='Phone_Number',inplace= True)
#df
```

```python
# now we should make the indexing from 0 to onward
df = df.reset_index(drop=True)
df
```
